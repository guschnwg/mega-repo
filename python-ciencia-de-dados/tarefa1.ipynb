Crie dois novos DataFrames, um contendo os registros de pessoas do sexo masculino e outra do sexo feminino. Use estes DataFrames para responder:

Qual é a proporção de reclamações entre homens e mulheres? (total de reclamações de homens dividido pelo total de reclamações de mulheres)

```python
# Resposta:
f = df.SEXO == 'F'
m = df.SEXO == 'M'

df_fem = df[f]
df_mas = df[m]

df_fem_recl = df_fem[df_fem.TIPO == 'RECLAMAÇÃO']
df_mas_recl = df_mas[df_mas.TIPO == 'RECLAMAÇÃO']

proporcao = len(df_mas_recl) / len(df_fem_recl) * 100

print(f"A cada 100 reclamações de mulheres, temos {proporcao:.2f} reclamações de homens")
```

Concatene os dois DataFrames criados acima para criar um DataFrame chamado `df_todos`. Verifique se este DataFrame tem o mesmo número de linhas do DataFrame lido do CSV.

```python
# Resposta:
df_todos = df_fem + df_mas
len(df_todos) == len(df)
```

Crie um novo DataFrame chamado `df_sexo` contendo a contagem de registros para cada sexo. A coluna com a contagem deve se chamar `contagem`.

```python
# Resposta:
df_sexo = df.groupby("SEXO").size().to_frame("contagem").sort_values("contagem", ascending=False)
df_sexo
```

Crie um novo DataFrame chamado `df_sexo_tipo` contendo a contagem de registros para cada sexo e tipo. A coluna com a contagem deve se chamar `total`.

```python
# Resposta:
df_sexo_tipo = df.groupby(["SEXO", "TIPO"]).size().to_frame("total").sort_values(["SEXO", "total"], ascending=False)
df_sexo_tipo
```

Faça uma junção (join ou merge) dos DataFrames `df_sexo, df_sexo_tipo` para criar um novo DataFrame chamado `df_juncao` contendo as contagens por sexo e também por sexo/tipo.

```python
# Resposta:
df_juncao = df_sexo.join(df_sexo_tipo)
df_juncao
```

Crie uma nova coluna no DataFrame `df_juncao` chamada `proporcao` contendo a porcentagem de homens e mulheres que fizeram cada tipo de registro. Responda: qual sexo faz mais elogios? Qual faz mais solicitações? Qual faz mais reclamações?

```python
# Resposta:

df_tipo = df.groupby("TIPO").size().to_frame("contagem_tipo").sort_values("contagem_tipo", ascending=False)
df_juncao = df_juncao.join(df_tipo)

df_juncao["proporcao"] = df_juncao["total"] / df_juncao["contagem"]
df_juncao["proporcao_tipo"] = df_juncao["total"] / df_juncao["contagem_tipo"]
df_juncao
```
